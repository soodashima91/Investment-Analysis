##### Creating a dataframe D1 with conditions like country as USA, funding_round_type = venture and raised amount between 5 million usd and 15 million usd. Also grouping of dataframe on main_sector and then printing them in descending order.

```python
D1 = master_frame[(master_frame['country_code']=='USA') & (master_frame['funding_round_type']=='venture')& (master_frame['raised_amount_usd']>=5000000) & (master_frame['raised_amount_usd']<=15000000)]
D1_group = D1.groupby(by='main_sector')
D1_total=D1_group[['raised_amount_usd']].sum().sort_values(by=['raised_amount_usd'], ascending=False)
D1_total
```

```python
D1_count=D1_group[['raised_amount_usd']].count().sort_values(by=['raised_amount_usd'], ascending=False)
D1_count
```

##### merging dataframes with sum and count of total investments

```python
D1_merge = D1_total.merge(D1_count, how='inner', on ='main_sector')
pd.DataFrame(D1_merge)
```

##### Renaming columns to Total_sum and Total_count of investments

```python
D1_merge=D1_merge.rename(columns={"raised_amount_usd_x": "Total_sum", "raised_amount_usd_y": "Total_count"})
D1_merge
```

```python
D1 = D1.merge(D1_merge, how='left', on='main_sector')
```

##### Finally D1 is created for USA country containing the observations of funding type FT falling within the 5-15 million USD range. The data frames contain all the columns of the master_frame along with the main sector and the primary sector.
